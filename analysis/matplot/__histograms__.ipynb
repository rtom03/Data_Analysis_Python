```python
import ast 
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../pandas/data_jobs.csv')

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])



def histograms(x):
    if pd.notna(x):
        return ast.literal_eval(x)
    else:
        return x
df['job_skills'] = df['job_skills'].apply(histograms)
fig,ax = plt.subplots()
df_DA = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')]
df_DA['salary_year_avg'].plot(kind='hist',bins=70,edgecolor='black')

plt.xlim(0,250000)
ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))
```
